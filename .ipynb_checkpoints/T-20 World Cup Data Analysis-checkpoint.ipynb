```python
import pandas as pd
import json
```

```python
file_path = "t20_json_files\\t20_wc_match_results.json"
with open(file_path) as f:
        data = json.load(f)
len(data)
data
```

```python
df_match=pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard':'match_id'},axis=1,inplace=True)
df_match.head()
```

```python
match_id_dict={} 
for index,row in df_match.iterrows():
    key1=row['team1'] +" Vs "+row['team2']
    key2=row['team2'] +" Vs "+row['team1']
    
    match_id_dict[key1]=row['match_id']
    match_id_dict[key2]=row['match_id']

match_id_dict
```

# Batting Details

```python
file_path = "t20_json_files\\t20_wc_batting_summary.json"
with open(file_path) as f:
        data = json.load(f)
        
        records=[]
        for rec in data:
            records.extend(rec['battingSummary'])
        
len(records)
```

```python
df_batting=pd.DataFrame(records)
df_batting.head()
```

```python
df_batting["out/not_out"]=df_batting.dismissal.apply(lambda x:"out" if len(x)>0 else "not_out")
df_batting.head()
```

```python
df_batting.drop(columns=["dismissal"],inplace=True)
```

```python
df_batting.head(15)
```

```python
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x:  x.replace('â€',''))
```

```python
df_batting.head(3)
```

```python
df_batting["match_id"]=df_batting["match"].map(match_id_dict)
df_batting.head()
```

```python
df_batting.to_csv("t20_csv//match_batting_summary",index=False)
```
